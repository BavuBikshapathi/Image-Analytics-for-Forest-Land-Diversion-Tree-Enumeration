```python
def count_bounding_boxes(image_path):
    # Perform prediction
    results = model.predict(source=image_path, conf=0.1)

    # Count the total number of bounding boxes detected
    total_bounding_boxes = len(results[0].boxes.data)
    print(f"Total number of trees detected: {total_bounding_boxes}")
    return total_bounding_boxes

# Specify the path to the test image
test_image_path = "/content/runs/detect/predict/000000_jpg.rf.915026bba24c92d05b9c2418de59a6ed.jpg"

# Count the total bounding boxes in the test image
count_bounding_boxes(test_image_path)
```
